## Checking for outliers in features

```python
outlierss=[]
def detectt_outlierss(data):
    quantile1, quantile3=np.percentile(data,[25,75])
    iqr_value=(quantile1-quantile3)
    lower_bound_val=quantile1-(1.5*iqr_value)
    upper_bound_val=quantile3+(1.5*iqr_value)
    for i in range (len(data)):
        if data[i]>upper_bound_val: 
            outlierss.append(data[i])
        elif data[i]<upper_bound_val:
            outlierss.append(data[i])
        return outlierss
```

```python
x55=df["X55"]
x56=df["X56"]
x57=df["X57"]
```

```python
outliers_pt55=detectt_outlierss(x55)
outliers_pt56=detectt_outlierss(x56)
outliers_pt57=detectt_outlierss(x57)
print(outliers_pt55)
print(outliers_pt56)
print(outliers_pt57)
```
